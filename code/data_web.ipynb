```python
import datetime
import requests
import zipfile
from io import BytesIO

# 定义链接的基本格式
base_url = "https://data.binance.vision/data/spot/daily/klines/BTCUSDT/1d/BTCUSDT-1d-{date}.zip"
# 定义日期范围
start_date = datetime.date(2017, 8, 17)  # 起始日期
end_date = datetime.date(2025, 1, 13)   # 结束日期

# 生成日期范围内的所有链接
links = []
current_date = start_date
while current_date <= end_date:
    # 格式化日期为YYYYMMDD格式
    formatted_date = current_date.strftime("%Y-%m-%d")
    # 生成链接
    link = base_url.format(date=formatted_date)
    links.append(link)
    # 增加一天
    current_date += datetime.timedelta(days=1)
    # if(current_date == end_date):
    #     print("XIBOXUI")
    # if(current_date == datetime.date(2025, 1, 14) ):
    #     print("LKXUI")

# 打印生成的链接

with open ("./datas/output.txt","w") as f:
    for link in links:
        f.write(link)
        f.write("\n")


# for link in links:
    
#     with open("output.txt","w") as f:
#         f.writelines(link)
#     # response = requests.get(link)
#     print("ZELA")
#     print(link)
    
    
    
#     # # response.raise_for_status()
#     # if response.status_code == 200:
#     #     print("Girl")
#     # # 将内容读取到内存中
#     #     zip_content = BytesIO(response.content)
    
#     # # 使用zipfile打开内存中的zip文件
#     #     with zipfile.ZipFile(zip_content, 'r') as zip_ref:
#     #     # 打印zip文件中的所有文件名
#     #        print(zip_ref.namelist())
        
#     #     # 假设我们知道zip文件中有一个名为'example.txt'的文件
#     #     # 我们可以提取并读取这个文件的内容
#     #        with zip_ref.open('example.txt') as file:
#     #         content = file.read()
#     #         print(content.decode('utf-8'))  # 假设文件是UTF-8编码的文本文件
# else:
#     print(f"Failed to retrieve the zip file: Status code {response.status_code}")
```
